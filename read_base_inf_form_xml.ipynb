```python
import xml.dom.minidom as dm
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
```

```python
df_data = pd.DataFrame()
signals_path = ["./HTN", "./NHTN"]
samp_num = 0
```

```python
df_data = pd.DataFrame()
num = 0
for path_ch in signals_path:
    seq_files = os.listdir(path_ch)#返回指定的文件夹包含的文件或文件夹的名字的列表
    seq_files.sort(key=lambda x:int(x.split('.')[0])) #按“.”分割，并把分割结果的[0]转为整形并排序      #排序
    if(path_ch == signals_path[0]):
        state = "HTN"
    else:
        state = "NHTN"
    for fil_ch in tqdm(seq_files):
        patienNumber = num
        doc = dm.parse(os.path.join(path_ch,fil_ch))
        try:
            name =  (doc.getElementsByTagName('name'))[1].childNodes[0].data
        except :
            name = "error"
        date = (doc.getElementsByTagName('center'))[0].getAttribute("value")
        gender =  (doc.getElementsByTagName('administrativeGenderCode'))[0].getAttribute("code")
        gread = (doc.getElementsByTagName('gread'))[0].childNodes[0].data
        digcode = (doc.getElementsByTagName('digcode'))[0].childNodes[0].data
        birthTime = (doc.getElementsByTagName('birthTime'))[0].getAttribute("value")
        data = {}
        data['patienNumber'] = patienNumber,
        data['name'] = name,
        data['state']= state,
        data['date']= date,
        data['birthTime'] = birthTime,
        data['gender'] = gender,
        data['gread'] = gread,
        data['digcode'] = digcode
        df_data = df_data.append(data,ignore_index=True)
        num = num + 1
```
