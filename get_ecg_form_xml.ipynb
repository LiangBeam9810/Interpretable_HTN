```python
import xml.dom.minidom as dm
import numpy as np
import os
from tqdm import tqdm
```

```python
EcgLength_num = 12000
EcgChannles_num = 12
data_path = ["./HTN", "./NHTN"] #路径
```

```python
samp_num = 1      #记录总样本数量 
list_lable = []  #记录标签

for path_i in data_path:
    print("check--"+path_i+"\n")
    seq_files = os.listdir(path_i)#返回指定的文件夹包含的文件或文件夹的名字的列表
    seq_files.sort(key=lambda x:int(x.split('.')[0])) #按“.”分割，并把分割结果的[0]转为整形并排序
    for file_i in (seq_files):

        print(str(file_i))
        if  (path_i == data_path[0]): #添加标签
            list_lable.append(1)
            samp_file_name = "./npy_ECG/"+str(samp_num)+"_HTN.npy"
        else :
            list_lable.append(0)
            samp_file_name = "./npy_ECG/"+str(samp_num)+"_NHTN.npy"

        xml_doc = dm.parse(os.path.join(path_i,file_i)) #打开该xml文件
        nope_root = xml_doc.getElementsByTagName('digits')#寻找名为digits的子类
        nda_samp_i = np.empty([EcgChannles_num,EcgLength_num], dtype = int)
        for channle_i in (range(EcgChannles_num)): 
            list_buf = nope_root[channle_i].childNodes[0].data.split(" ") ##第i导联数据第i组序列中 取出text序列，并以‘ ’分割，得到list
            len_of_buf = len(list_buf)
            list_buf = list_buf[:len_of_buf-1] #去掉最后一个空位
            len_of_buf = len(list_buf)
            print('Channel: '+str(channle_i)+" Length: "+str(len_of_buf))
            for j in range(EcgLength_num):        #取前int_sequence_num个数据点
                if(j > (len_of_buf - 1)):         #过短时补零
                    nda_samp_i[channle_i,j] = 0
                else:
                    nda_samp_i[channle_i,j] = eval(list_buf[j]) #转化为数值型
        np.save(samp_file_name,nda_samp_i)
        samp_num +=1
nda_lable = np.array(list_lable)
print(type(list_lable))
print(type(nda_lable))
print ("the number of sample:",samp_num)
print("the size of lable:",nda_lable.size)
print("the number of hypertension:",(nda_lable[nda_lable == 1]).size)
print("the number of normal:",(nda_lable[nda_lable == 0]).size)
np.save("./label.npy",nda_lable)
```
